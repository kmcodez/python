thread = a flow of execution.Like a separate order of instructions,<br>
          However each thread takes a turn running to achieve concurrency<br>
          GIL = (global interpreter lock)<br>
          allows only one thread to hold the control of the python interpreter

cpu bound = program/task spends most of it's time waiting for internal events (cpu intensive)<br>
              use multiprocessing

io bound = program/task spends most of it's time waiting for external events (user input, web scraping)<br>
              use multiprocessing

```python
import threading
import time
```

# Return the number of Thread objects currently alive.<br>
print(threading.active_count())<br>
# Return a list of all Thread objects currently alive.<br>
print(threading.enumerate())

```python
def eat_breakfast():
    time.sleep(3)
    print('you ate breakfast')
```

```python
def drink_coffee():
    time.sleep(4)
    print('you drank coffee')
```

```python
def study():
    time.sleep(5)
    print('you finish studying')
```

cost 12 seconds

```python
eat_breakfast()
drink_coffee()
study()
```

create a thread

```python
x = threading.Thread(target=eat_breakfast, args=())
x.start()
```

create a thread

```python
y = threading.Thread(target=drink_coffee, args=())
y.start()
```

create a thread

```python
z = threading.Thread(target=study, args=())
z.start()
```

Wait until the thread terminates.

```python
x.join()
y.join()
z.join()
```
