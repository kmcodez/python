reduce() = apply a function to an iterable and reduce it to a single cumulative value<br>
          performs function on first two elements and repeats process until 1 value remains<br>
<br>
reduce(function, iterable)

```python
import functools
```

process<br>
step1:  ["HE", "L", "L", "O"]<br>
step2:  ["HEL", "L", "O"]<br>
step3:  ["HELL", "O"]<br>
step4:  ["HELLO"]

```python
letters = ["H", "E", "L", "L", "O"]
word = functools.reduce(lambda first, second: first + second, letters)
print(word)
```

```python
factorial = [5, 4, 3, 2, 1]
result = functools.reduce(lambda first, second: first * second, factorial)
print(result)
```
