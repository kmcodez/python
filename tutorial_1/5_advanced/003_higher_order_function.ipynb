Higher Order Function = a function that either:<br>
                          1. accepts a function as argument<br>
                                  or<br>
                          2. returns a function<br>
                          (in python, functions are also treated as objects)

case 1

```python
def loud(text):
    return text.upper()
```

```python
def quiet(text):
    return text.lower()
```

```python
def hello(function):
    text = function("HeLlo")
    print(text)
```

```python
hello(loud)
hello(quiet)
```

case 2<br>
just like the closure function

```python
z = 10
```

```python
def divisor(x):
    def dividend(y):
        return y / x * z
    return dividend
```

```python
divide = divisor(2)
print(divide(10))
```
