daemon thread = a thread runs in the background, not important for programs to run<br>
                your program will not wait for daemon to complete before exiting<br>
                non-daemon threads cannot normally be killed, stay alive until task is complete<br>
<br>
                ex. background tasks, garbage collection,waiting for input, long running processes

```python
import threading
import time
```

```python
def timer():
    print()
    count = 0
    while True:
        time.sleep(1)
        count += 1
        print('logged in for: ', count, 'seconds')
```

a daemon thread can be killed automatically

```python
x = threading.Thread(target=timer)
x.start()
```

```python
x = threading.Thread(target=timer, daemon=True)
x.start()
```

```python
x.setDaemon(True)
print(x.isDaemon())
```
