lambda function = function written in 1 line using lambda word<br>
                  accepts any number of arguments, but only has one expression<br>
                  (think it as a shortcut)<br>
                  (useful if needed for a short period of time, throw-away)<br>
lambda parameters:expression

case 1

```python
def double(x):
    return x * 2
```

```python
print(double(10))
```

# case 2<br>
if you have learned js, it looks like Arrow function expressions double = x => x * 2

```python
double = lambda x: x * 2
print(double(5))
```

```python
multiply = lambda x, y: x * y
print(multiply(3, 4))
```

```python
add = lambda x, y, z: x + y + z
print(add(1, 2, 3))
```

```python
check_age = lambda age: True if age >= 18 else False
print(check_age(3))
```

```python
full_name = lambda first_name, last_name: first_name + " " + last_name
print(full_name("Bro", "Code"))
```
