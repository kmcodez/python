multiple inheritance = when a class is derived from more than one parent class

```python
class Prey:
    def flee(self):
        print("This animal flees")
```

```python
class Predator:
    def hunt(self):
        print("This animal is hunting")
```

```python
class Rabbit(Prey):
    pass
```

```python
class Hawk(Predator):
    pass
```

```python
class Fish(Prey, Predator):  # mulptiple inheritance
    pass
```

```python
rabbit = Rabbit()
hawk = Hawk()
fish = Fish()
```

```python
rabbit.flee()
hawk.hunt()
fish.flee()
fish.hunt()
```
