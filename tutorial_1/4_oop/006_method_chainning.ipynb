method chaining = calling multiple methods sequentially.<br>
                  each call performs an action on the same objects and returns self

```python
class Car:
    def turn_on(self):
        print("You start the engine")
        return self
    def drive(self):
        print("You drive the car")
        return self
    def brake(self):
        print("You step on the brakes")
        return self
    def turn_off(self):
        print("You turn off the engine")
        return self
```

```python
car = Car()
```

```python
car.turn_on().drive().brake().turn_off()
car.turn_on()\
    .drive()\
    .brake()\
    .turn_off()
```
