```python
class Car:
    color = None
```

```python
class Motorcycle:
    color = None
```

```python
def change_color(car_intance, color):
    car_intance.color = color
```

```python
car_1 = Car()
car_2 = Car()
car_3 = Car()
```

```python
bike_1 = Motorcycle()
bike_2 = Motorcycle()
bike_3 = Motorcycle()
```

```python
change_color(car_1, "red")
change_color(car_2, "blue")
change_color(car_3, "green")
change_color(bike_1, "black")
```

```python
print(car_1.color)
print(car_2.color)
print(car_3.color)
print(bike_1.color)
```
