Duck typing = concept where the class of an object is less important than the methods / attributes<br>
              class type is not checked if minimum methods / attributes are present<br>
              "if it walks like a duck, and it quacks like a duck, then it must be a duck"

```python
class Duck:
    def walk(self):
        print("this duck is walking")
    def talk(self):
        print("this duck is quacking")
```

```python
class Chicken:
    def walk(self):
        print("this chicken is walking")
    def talk(self):
        print("this chicken is clucking")
```

```python
class Person:
    def catch(self, duck):
        duck.walk()
        duck.talk()
        print("You caught the critter")
```

```python
duck = Duck()
chicken = Chicken()
person = Person()
```

```python
person.catch(duck)
```

because the Chicken class has the similar attributes / methods as the Duck class<br>
so you can also pass chicken instance to the function

```python
person.catch(chicken)
```
