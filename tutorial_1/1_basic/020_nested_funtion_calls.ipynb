nested function calls = function calls inside other function calls<br>
                        innermost function calls are resolved first<br>
                        returned value is used as argument for the next outer function call

```python
num = input("Enter a whole positive number: ")
num = float(num)
num = abs(num)
num = round(num)
print(num)
```

```python
print(round(abs(float(input("Enter a whole positive number: ")))))
```
