function = a block of code which is executed only when it is called

```python
def hello(first_name, last_name, age):
    print("hello " + first_name + " " + last_name)
    print("You are " + str(age) + " years old")
    print("have a nice day!")
```

return statement =  Functions send Python values/objects back to the caller.<br>
                    These values/objects are known as the function's return value

```python
def multiply(number1, number2):
    return number1 * number2
```

```python
x = multiply(6, 8)
# print(x)
```

keyword arguments = arguments preceded by an identifier when we pass them to a function<br>
                    The order of the arguments does not matter, unlike positional arguments<br>
                    Python knows the names of the argumetns that our function receives.

```python
def hello_2(first: str, middle: str, last: str):
    print(first, middle, last)
```

call the function through passing the keyword

```python
hello_2(last="Code", middle="Dude", first="Bro")
```
