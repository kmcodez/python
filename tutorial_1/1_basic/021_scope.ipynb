scope = The region that a variable is recognized<br>
        A variable is only available from inside the region it is created<br>
        A global and locally scoped versions of a variable can be created

python used LEGB rule for following<br>
Local, Enclosing, Global, Built-in

```python
name = "Bro"    # global scope (available anywhere in the code)
```

```python
def display_name():
    name = "Code"   # local scope (available only inside the function)
    print(name)
```

```python
display_name()
print(name)
```
