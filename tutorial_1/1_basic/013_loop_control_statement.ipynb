Loop control statement = change a loops execution from its normal sequence

break     = used to terminate the loop entirely<br>
continue  = skips to the next iteration of the loop<br>
pass      = does nothing, acts a placeholder

break

```python
while True:
    name = input("Enter your name: ")
    if name != "":
        break
```

continue

```python
phone_number = "123-456-7890"
for i in phone_number:
    if i == "-":
        continue
    print(i, end="")
```

pass

```python
for i in range(1, 21):
    if i == 13:
        pass
    else:
        print(i)
```
