dictionary    = a changeable, unordered collection of unique key:value pairs<br>
              Fast because they use hashing, allow us to access a value quickly

```python
capital = {
    'USA': "Washington DC",
    "India": "New Delhi",
    "China": "Beijing",
    "Russia": "Moscow",
}
```

update

```python
capital["Germany"] = "Berlin"
capital.update({"Germany": 'Berlin'})
```

remove

```python
capital.pop('China')
```

remove all

```python
capital.clear()
```

```python
print(capital['Germany'])
print(capital.get('Germany'))
print(capital.keys())
print(capital.values())
print(capital.items())
```

```python
for key, value in capital.items():
    print(key, ":", value)
```
