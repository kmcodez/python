```python
import os  # os includes the standard python modules
import shutil  # includes the module for copying files
```

```python
path = "C:\\Users\\STRATAZON\\Downloads\\learning codez\\codez\\python\\Bro_Code_Python\\0_files\\test.txt"
```

print('---------------------{:^20}----------------------'.format( "check if a file exists"))<br>
# check if a file exists and the type of the file(file / dir?)<br>
if os.path.exists(path):<br>
    print("that location exists")<br>
    if os.path.isfile(path):<br>
        print("that is a file")<br>
    elif os.path.isdir(path):<br>
        print("that is a folder")<br>
else:<br>
    print("that location does not exist")

print('---------------------{:^20}----------------------'.format( "open and read a file"))<br>
# open and read a file, if you add "with", it will close the file automatically<br>
try:<br>
    with open(path) as file:<br>
        print(file.read())<br>
except FileNotFoundError as e:<br>
    print(e)<br>
    print("That file was not found")<br>
except Exception as e:<br>
    print(e)<br>
    print("something went wrong :(")

# it will print true, which means the file has been already closed.<br>
# print(file.closed)

print('---------------------{:^20}----------------------'.format("open then write a file"))<br>
# open then write a file<br>
# the second argument is 'r' which means 'read the file' by default, 'w' means 'overwrite the file', 'a' means 'append the file'

with open(path, 'w') as file:<br>
    file.write("Yoooooooo\nThis is some text\nHave a good one!\n")<br>
with open(path, "a") as file:<br>
    file.write("Have a nice day! see you!")

print('---------------------{:^20}----------------------'.format("copy files"))<br>
# copy files

# copyfile()    = copies contents of a file<br>
# copy()        = copyfile() + permission mode + destination can be a directory<br>
# copy2()       = copy() + metadata (file's creation and modification times)

# of course, you can use copy and copy2 methods to replace the copyfile method even without changing the parameters<br>
