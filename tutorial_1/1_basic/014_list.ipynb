list = used to store multiple items in a single valuable

```python
food = ["pizza", "hamburger", "hotdog", "spaghatti", "pudding"]
```

```python
food[0] = "sushi"
```

```python
food.append("ice cream")
food.remove("hotdog")
food.pop()
food.insert(3, "cake")
food.sort()
food.clear()
```

```python
for item in food:
    print(item)
```
