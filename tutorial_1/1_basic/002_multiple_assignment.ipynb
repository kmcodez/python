multiple assignments = allows us to assign multiple variables at the same time in one line of code

```python
name = "Bro"
age = 21
attractive = True
print(name)
print(age)
print(attractive)
```

```python
name, age, attractive = "Bro", 21, True
print(name)
print(age)
print(attractive)
```

```python
spongebob = 30
patrick = 30
sandy = 30
squidward = 30
print(spongebob)
print(patrick)
print(sandy)
print(squidward)
```

```python
spongebob = patrick = sandy = squidward = 30
print(spongebob)
print(patrick)
print(sandy)
print(squidward)
```

```python
# Assigning multiple values to multiple variables
a, b, c = 5, 3.2, "Hello"

print (a)
print (b)
print (c)
```

```python
x = y = z = "same"

print (x)
print (y)
print (z)
```
