index operator [] = give access to a sequence's elements (set, list, tuple)<br>
                      by their position in the sequence

```python
name = "bro Code!"
```

```python
if name[0].islower():
    name = name.capitalize()
```

```python
print(name)
```

```python
first_name = name[:3].upper()
last_name = name[4:].lower()
last_character = name[-1]
```

```python
print(first_name)
print(last_name)
print(last_character)
```
